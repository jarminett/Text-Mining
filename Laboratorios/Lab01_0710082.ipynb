Realice una comparación entre el algoritmo de Canny, el operador de Sobel y el operador Laplaciano para encontrar los bordes en una imagen, para este ejercicio puede utilizar las funciones de openCV y cualquier imagen, como entregable debe mostrar una comparación entre los tres enfoques con una tabla donde se muestre las ventajas y desventajas de cada enfoque.


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
input_image = cv2.imread("nick.jpg", cv2.IMREAD_COLOR)
input_image_gray = cv2.cvtColor(input_image,cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(7,7))
plt.imshow(cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB))
plt.show()
```

**Algoritmo de Canny**

Es un algoritmo desarrollado por John F. Canny, de ahí su nombre, el cual es bastante utilizado para la detección de bordes en imágenes debido a que es menos sensible al ruid comparado con el operador Laplaciano debido a que uno de los pasos del algoritmo se basa en aplicar un filtro Gaussiano a la imagen previo a la detección de bordes.

```python
plt.figure(figsize=(6,6))
plt.imshow(cv2.Canny(input_image_gray,100,200),cmap='gray')
plt.show()
```

**Operador de Sobel**

Este operador se basa en identificar la intensidad del gradiente entre 2 puntos de la imagen, con el fin de identificar un posible borde entre un pixel y otro. El operador de sobel añade un pequeño suavizado a la imagen con el fin de minimizar la detección de bordes falsos a causa del ruido.

```python
sobel_x = cv2.Sobel(input_image_gray, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(input_image_gray, cv2.CV_64F, 0, 1, ksize=3)
abs_sobel_x = cv2.convertScaleAbs(sobel_x)
abs_sobel_y = cv2.convertScaleAbs(sobel_y)
sobel = cv2.addWeighted(abs_sobel_x, 0.5, abs_sobel_y, 0.5, 0)
plt.figure(figsize=(6,6))
plt.imshow(sobel,cmap='gray')
plt.show()
```

**Laplaciano**

El filtro Laplaciano también es una máscara que se opera sobre la imagen, la ventaja de este a diferencia del de Sobel es que solo se debe de operar una vez, sin embargo, este tipo de operador es muy sensible a ruido, llegando a identificar bordes falsos, por lo cual se recomienda que antes de aplicar el operador Laplaciano, se aplique un filtro gaussiano a la imagen para eliminar le ruido.

```python
plt.figure(figsize=(6,6))
plt.imshow(cv2.Laplacian(input_image_gray,cv2.CV_8U), cmap='gray')
plt.show()
```
